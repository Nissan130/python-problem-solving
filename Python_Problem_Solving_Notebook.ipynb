**Problem 1: Count Vowels and Consonants**

Concepts: strings, loops, conditionals  
Problem Statement: Given a string, count how many vowels and consonants it contains.
Ignore spaces and non-alphabet characters.  
Vowels: a, e, i, o, u (both upper & lower case) 

Sample Input:  
Hello World     
Sample Output:   
Vowels: 3   
Consonants: 7   

```python
#problem-1 solution
def vowel_consonant_count(text):
    vowel_count = 0
    consonant_count = 0
    vowels = ['a','e','i','o','u']

    for ch in text:
        if 'a'<=ch<='z' and ch not in vowels:
            consonant_count +=1
        elif ch in vowels:
            vowel_count+=1
        else:
            continue
    return vowel_count, consonant_count

input_text = input("Enter the string: ").lower()

vowel, consonant = vowel_consonant_count(input_text)

print("Your Text:", input_text)
print("Vowels: ", vowel)
print("Consonants: ", consonant)
```

**Problem 2: Reverse Words in a Sentence**  

Concepts: strings, split/join, loops  
Task:  
Given a sentence, reverse the order of words.  
Sample Input:  
I love Python programming  

Sample Output  
programming Python love I


```python
# problem-2 solution

def reverse_sentence_words(text):
    splited_words = text.split()
    
    for i in range(len(splited_words)-1,-1,-1):
        print(splited_words[i],end=' ')
    

sentence = input("Enter a sentence: ")

print("Input: ",sentence)
print("Output: ", end=' ')
reverse_sentence_words(sentence)
```

**Problem 3: Count Character Frequency**       

Concepts: dictionaries, loops   
Task:   
Given a string, count how many times each character appears (ignore spaces).    

Sample Input:      
banana

Sample Output:  
{'b': 1, 'a': 3, 'n': 2}

```python
#problem-3 solution

def count_character_frequency(text):
    character_dict = {}
    for ch in text:
        if ch == ' ':
            continue
        elif ch not in character_dict:
            character_dict[ch] = [ch]
        else:
            character_dict[ch].append(ch)
    
    #calculate the length of each key
    character_frequency = {}
    for key in character_dict:
        character_frequency[key] = len(character_dict[key])
    return character_frequency
        
text = input("Enter a text: ")
print("Input: ",text)
print("Output: ")
count_character_frequency(text)
```

**Problem 4: Find Second Largest Number**   

Concepts: lists, sorting, conditionals  
Task:   
Given a list of integers, find the second largest unique number.    
Sample Input:      
5 1 9 3 9 7     
Sample Output:  
7

```python
#problem-4 solution

def find_second_largest_number(num_list):
    num_list = list(dict.fromkeys(num_list))  #remove the duplicate numbers
    max = num_list[0]
    max_index = 0
    for i in range(len(num_list)):
        if num_list[i] > max:
            max = num_list[i]
            max_index +=1
    num_list.remove(max)

    #find the second largest value
    second_max = num_list[0]
    for i in range(len(num_list)):
        if num_list[i] > second_max:
            second_max = num_list[i]
    return second_max

num_list = list(map(int, input().split()))
print("Input: ", num_list)
print('Second largest number is: ', find_second_largest_number(num_list))

```

**Problem 5: Check for Palindrome (Ignoring Case)**     

Concepts: strings, slicing, conditionals    
Task:   
Check whether a given string is a palindrome (case-insensitive).    
Sample Input:   
Madam   
Sample Output:  
True

```python
#solution problem-5
def check_palindrome(text):
    text = text.lower()
    reversed_text = ""
    for i in range(len(text)-1,-1,-1):
        reversed_text += text[i]
    if reversed_text == text:
        return True
    else:
        return False 

text = input("Enter a text: ")
print("Input: ",text)
print("Output: ", check_palindrome(text))
```

**Problem 6: Word Frequency Counter**   

Concepts: dictionaries, text processing 
Task:   
Given a sentence, print the frequency of each word.      
Sample Input:   
python is fun and python is powerful    

Sample Output:  
{'python': 2, 'is': 2, 'fun': 1, 'and': 1, 'powerful': 1}

```python
#solution problem-6
def word_frequency_counter(sentence):
    word_list = sentence.split()
    word_frequency_dict = {}
    for word in word_list:
        if word == '':
            continue 
        elif word not in word_frequency_dict:
            word_frequency_dict[word] = [word]
        else:
            word_frequency_dict[word].append(word)

    #count the word frequncy in each list
    word_frequency = {}
    for key in word_frequency_dict:
        word_frequency[key] = len(word_frequency_dict[key])
    return word_frequency

sentence = input("Enter a sentence: ")
print("Input: ",sentence)
print("Output: ")
word_frequency_counter(sentence)
```
